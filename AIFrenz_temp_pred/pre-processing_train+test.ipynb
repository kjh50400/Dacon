```python
import pandas as pd
from IPython.display import display
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
from matplotlib import gridspec
%matplotlib inline

from matplotlib import font_manager, rc
font_path='c:/Windows/Fonts/malgun.ttf'
font_name= font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
DIR_DATA = './AIFrenz_Season1_dataset/'
DIR_TRAIN = DIR_DATA + 'train.csv'
DIR_TEST = DIR_DATA + 'test.csv'
```

```python
train = pd.read_csv(DIR_TRAIN)

print(train.shape)
display(train.head())
```

```python
test = pd.read_csv(DIR_TEST)

print(test.shape)
display(test.head())
```

```python
Temperature = ['X00', 'X07', 'X28', 'X31', 'X32']
localP = ['X01', 'X06', 'X22', 'X27', 'X29']
windV = ['X02', 'X03', 'X18', 'X24', 'X26']
rain = ['X04', 'X10', 'X21', 'X36', 'X39']
seaP = ['X05', 'X08', 'X09', 'X23', 'X33']
solar = ['X11', 'X14', 'X16', 'X19', 'X34']
humi = ['X12', 'X20', 'X30', 'X37', 'X38']
windD = ['X13', 'X15', 'X17', 'X25', 'X35']
sensor = ['Y00', 'Y01', 'Y02', 'Y03', 'Y04',
          'Y05', 'Y06', 'Y07', 'Y08', 'Y09', 
          'Y10', 'Y11', 'Y12', 'Y13', 'Y14', 
          'Y15', 'Y16', 'Y17']
```

```python
X_train = train.loc[:, 'X00':'X39']

print(X_train.shape)
display(X_train.head())
```

```python
fig = plt.figure(figsize=(30,10))
spec = gridspec.GridSpec(ncols=2, nrows=1, width_ratios=[1,2])
ax1 = fig.add_subplot(spec[0])
ax2 = fig.add_subplot(spec[1])

ax1.plot(X_train[Temperature])
ax1.set_ylim(5, 40)
ax1.set_xlim(0, 144*33)
ax1.grid(which='both', axis='both')

ax2.plot(test[Temperature])
ax2.set_ylim(5, 40)
ax2.set_xlim(0, 144*80)
ax2.grid(which='both', axis='both')

plt.show()
```
