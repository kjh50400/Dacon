```python
temp = ['X00', 'X07', 'X28', 'X31', 'X32']
localP = ['X01', 'X06', 'X22', 'X27', 'X29']
windV = ['X02', 'X03', 'X18', 'X24', 'X26']
rain = ['X04', 'X10', 'X21', 'X36', 'X39']
seaP = ['X05', 'X08', 'X09', 'X23', 'X33']
solar = ['X11', 'X14', 'X16', 'X19', 'X34']
humi = ['X12', 'X20', 'X30', 'X37', 'X38']
windD = ['X13', 'X15', 'X17', 'X25', 'X35']
sensor = ['Y00', 'Y01', 'Y02', 'Y03', 'Y04',
          'Y05', 'Y06', 'Y07', 'Y08', 'Y09', 
          'Y10', 'Y11', 'Y12', 'Y13', 'Y14', 
          'Y15', 'Y16', 'Y17']
```

```python
X_train = train.loc[:, 'X00':'X39']

print(X_train.shape)
display(X_train.head())
```

```python
X_train[solar].plot(figsize=(20,10))
plt.title('트레이닝셋 누적일사량')
plt.grid(which='both', axis='x')
plt.xlim(144*32, 144*33)
```

```python
def momentValue(df, columns):
    for column in columns:
        temp_df = pd.DataFrame(df[column])
        temp_df['dummy'] = 0
        
        for i in range(1, temp_df.shape[0]):
            a = temp_df[column].iloc[i] - temp_df[column].iloc[i-1]
            if a > 0:
                temp_df['dummy'].iloc[i] = a
        
        df[column] = temp_df['dummy']
    
    return df
```

```python
moment_X_train = X_train.copy()

momentValue(moment_X_train, solar)

print(moment_X_train.shape)
display(moment_X_train.head())
```

```python
moment_X_train[solar].plot(figsize=(20,10))
plt.title('트레이닝셋 일사량')
plt.grid(which='both', axis='x')
plt.xlim(144*31-12, 144*31+12)
```

```python
fig=plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)


ax1.plot(X_train.index, X_train[solar[0]])
# ax1.set_xlim(600, 700)

ax2.plot(moment_X_train.index, moment_X_train[solar[0]])
# ax2.set_xlim(600, 700)
```
